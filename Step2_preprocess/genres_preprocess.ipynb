# Action drama is redundant

```python
df[(df["action drama"] == 1) & (df["action"] == 0)]
```

```python
df[(df["action drama"] == 1) & (df["drama"] == 0)]
```

```python
df.drop("action drama", axis=1, inplace=True)
```

```python
df.head()
```

# Genres with too few anime

```python
to_be_dropped = []
for column in df.columns:
    if len(df[(df[column] == 1)]) <= 3 and column != "title":
        to_be_dropped.append(column)
```

```python
len(to_be_dropped)
```

```python
print(to_be_dropped)
```

```python
df.drop(to_be_dropped, axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df.columns
```

# Repeated Attributes

## Magical girl

```python
len(df[(df["magical girl"] == 1)])
```

```python
len(df[(df["mahou shoujo"] == 1)])
```

```python
len(df[(df["magical girl"] == 1) & (df["mahou shoujo"] == 1)])
```

```python
new_col = np.logical_or(df["magical girl"], df["mahou shoujo"])
```

```python
df["magical_girl"] = new_col
```

```python
len(df[(df["magical_girl"] == 1)])
```

```python
df.drop(["magical girl", "mahou shoujo"], axis=1, inplace=True)
```
